```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
wine_data = pd.read_csv('wine.csv')
wine_data.head(10)
```

```python
wine_data.describe()
```

```python
wine_data.info()
```

```python
wine_new = wine_data.iloc[:,1:]
wine_new.head(10)
```

```python
from sklearn.preprocessing import scale
wine_norm = scale(wine_new)
wine_norm
```

```python
wine_new.corr()
```

```python
correlation = wine_new.corr()
plt.figure(figsize=(8,8))
sns.heatmap(correlation,cmap='viridis',square=True)
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
wine_new.shape
```

```python
pca = PCA(n_components=13)
```

```python
pca_value = pca.fit(wine_norm)
```

```python
var = pca.explained_variance_ratio_
var
```

```python
plt.figure(figsize=(6,4))
plt.bar(range(13), var, alpha=0.5, align='center',
        label='individual explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
```

```python
pca_new = PCA(n_components=3)
pca_values_new = pca_new.fit_transform(wine_norm)
```

```python
var1 = pca_new.explained_variance_ratio_
var1
```

```python
x_pca = pca_new.transform(wine_norm)
x_pca.shape
```

```python
plt.figure(figsize=(15,8))
plt.scatter(x_pca[:,0],x_pca[:,1],c='r')

#ax = plt.axes(projection='3d')
#ax.scatter3D(x_pca[:,0],x_pca[:,1], x_pca[:,2], c=x_pca[:,2], cmap='Greens')
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
```

```python
z = linkage(x_pca,method='complete', metric='euclidean')
plt.figure(figsize=(15,8))
sch.dendrogram(z,leaf_rotation=0)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
```

```python
h_cluster = AgglomerativeClustering(n_clusters=3,linkage='complete',affinity='euclidean').fit(x_pca)
cluster_label = pd.Series(h_cluster.labels_)
```

```python
wine_data['Cluster Labels'] = cluster_label
wine_data
```

```python
wine_data['Cluster Labels'].unique()
```

```python
from sklearn.cluster import KMeans 
```

```python
wcss=[]
for i in range(1,13):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0)
    kmeans.fit(x_pca)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,13),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Fitting K-MEans to the dataset
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=0)
y_kmeans=kmeans.fit_predict(x_pca)

#Visualize the clusters

plt.scatter(x_pca[y_kmeans==0,0],x_pca[y_kmeans==0,1],s=100,c='red',label='Cluster1')
plt.scatter(x_pca[y_kmeans==1,0],x_pca[y_kmeans==1,1],s=100,c='blue',label='Cluster2')
plt.scatter(x_pca[y_kmeans==2,0],x_pca[y_kmeans==2,1],s=100,c='green',label='Cluster3')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c='white',label='Centroids')

plt.title('Clusters of Wines')
plt.xlabel('Wines')
plt.ylabel('Categories')
plt.legend()
plt.show()
```
